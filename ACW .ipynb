### 1.2  Evaluation and Hyperparameters 

i) KFold and RepeatedStratifiedKFold:
According to (Venkata, 2018) RepeatedStratifiedKFold is a method that repeats Stratified K-Fold Cross validator n times with different randomisation in each iteration. The entire dataset is divided into a subset with an equal percentage of samples of each target class while KFold helps in resolving problems caused by under-fitting when removing part of the data to train our model. 

While our data having various attributes correlated to the risk, it is important to have an even percentage split to improve consistency in model prediction.  By reducing the training dataing data, there is a risk of losing important patterns/trends in the dataset which increases errors through induced bias. Based on this circumstances KFold cross validation is a suitable method  which provides ample data for training and validating the model.


ii) GridSearchCV: 

When training our models on a dataset, GridSearchCv helps select the model with the best performance, which is also known as hyperparameter tuning. The performances of a model depend on the set values for the hyperparameters. Manually executing this can be time-consuming; thus, the use of GridSearchCV to automate the tuning is used. This function helps to loop through predefined hyperparameters.  Also fit your model on the training set and use the best parameters from the listed hyperparameters to predict your test set. examples of the parameter that can be passed in:

```python
def model_performance(model_name, X_train, y_train, y_test, y_pred, model_y_train_pred):
    
    print ('Model name: %s'%model_name) # print model name
    print ('Roc Auc Score For y test: %f'%roc_auc_score(y_test, y_pred)) # display Roc score: reference section 1.1(b)
    print ('Roc Auc Score For y Train: %f'%roc_auc_score(y_train, model_y_train_pred))

    
    #generate a report for the precision, recall and F1_score: reference section 1.1(b)
    print(classification_report(y_test, y_pred)) 
    
    # roc understanding : reference section 1.1(b)
    false_positive_rate, true_positive_rate, thresholds = metrics.roc_curve(y_test, y_pred)
    roc_auc = metrics.auc(false_positive_rate, true_positive_rate)
    
    # plot classification TP and FP rate. 
    plt.title('Receiver Operating Characteristic')
    plt.plot(false_positive_rate, true_positive_rate, 'b',
    label='AUC = %0.2f'% roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.xlim([-0.1,1.2])
    plt.ylim([-0.1,1.2])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    # display the model confusion matrix:
    skplt.metrics.plot_confusion_matrix(y_test, y_pred)
    plt.show()
    
    # print  confision matrix values
    confusion_matrix = skplt.metrics.confusion_matrix(y_test, y_pred)
    TP = confusion_matrix[0][0]
    FP = confusion_matrix[0][1]
    FN = confusion_matrix[1][0]
    TN = confusion_matrix[1][1]
    
    print ('True Positive: %s'%confusion_matrix[0][0])
    print ('False Positive: %s'%confusion_matrix[0][1])
    print ('False Negative: %s'%confusion_matrix[1][0])
    print ('True Negative: %s'%confusion_matrix[1][1])
    
    
```

```python
# https://scikit-learn.org/stable/auto_examples/feature_selection/plot_feature_selection.html#sphx-glr-auto-examples-feature-selection-plot-feature-selection-py
def feature_selection():
    
    print(__doc__)

    # #############################################################################
    # Import some data to play with

    columns =['IPSI' ,  'Contra', 'AF', 'CVA', 'TIA', 'ASx', 'Diabetes', 'IHD' , 'Hypertension', 'Arrhythmia', 'History']
    # create a new data frame with the specified columns
    column = cleaned_dataset[columns] # 
    
    # split into X and y train set
    X = cleaned_dataset[columns]
    y = cleaned_dataset['Risk']

    # Some noisy data not correlated
    E = np.random.RandomState(42).uniform(0, 0.1, size=(X.shape[0], 20))

    # Add the noisy data to the informative features
    X = np.hstack((X, E))

    # Split dataset to select feature and evaluate the classifier
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)

    plt.figure(1)
    plt.clf()

    X_indices = np.arange(X.shape[-1])

    # #############################################################################
    # Univariate feature selection with F-test for feature scoring
    # We use the default selection function to select the four
    # most significant features
    # print attributes and scores
    selector = SelectKBest(f_classif, k=10)
    selector.fit(X_train, y_train)
    single_feature_scores = pd.DataFrame(selector.scores_)
    selected = pd.DataFrame(column.columns)
    feature_scores = pd.concat([selected, single_feature_scores], axis=1)
    feature_scores.columns = ['Attributes', 'Score']
    print(feature_scores.nlargest(9, 'Score'))
    scores = -np.log10(selector.pvalues_)
    scores /= scores.max()
    plt.bar(X_indices - .45, scores, width=.2,
            label=r'Univariate score ($-Log(p_{value})$)')

    # #############################################################################
    # Compare to the weights of an SVM
    clf = make_pipeline(MinMaxScaler(), LinearSVC())
    clf.fit(X_train, y_train)
    print('Classification accuracy without selecting features: {:.3f}'
          .format(clf.score(X_test, y_test)))

    svm_weights = np.abs(clf[-1].coef_).sum(axis=0)
    svm_weights /= svm_weights.sum()

    plt.bar(X_indices - .25, svm_weights, width=.2, label='SVM weight')

    clf_selected = make_pipeline(
            SelectKBest(f_classif, k=10), MinMaxScaler(), LinearSVC()
    )
    clf_selected.fit(X_train, y_train)
    print('Classification accuracy after univariate feature selection: {:.3f}'
          .format(clf_selected.score(X_test, y_test)))

    svm_weights_selected = np.abs(clf_selected[-1].coef_).sum(axis=0)
    svm_weights_selected /= svm_weights_selected.sum()

    plt.bar(X_indices[selector.get_support()] - .05, svm_weights_selected,
            width=.2, label='SVM weights after selection')


    plt.title("Comparing feature selection")
    plt.xlabel('Feature number')
    plt.yticks(())
    plt.axis('tight')
    plt.legend(loc='upper right')
    plt.show()
```

### 8.1(c) Model Evaluation

In the code below, the kfold split  is used with 6 splits. This trains the model on 6 different splits and provide the mean and standard deviation of the model. from the score below, it can be seen that 4 model performed better with the train.Overfitting and underfitting can be detected when a model is applied from the output of a model. The following methodologies can be used to avoid overfitting. Underfitting is the process where a model is not trained with high precison. A cross validation score will be used to analyse the train data. In the code below, an array of rasult and name to hold the name and score of the model evaluation and using seed to maintain consistency. 

```python
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=6, random_state=seed) #  run Kfold to split data and 
    cv_results = model_selection.cross_val_score(model, X_train, y_train,
    cv=kfold, scoring=scoring)
    results.append(cv_results) # append result 
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    #source: (GitHub, n.d.)
```
